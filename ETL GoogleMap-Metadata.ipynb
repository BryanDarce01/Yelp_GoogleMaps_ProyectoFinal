```python
import pandas as pd
import os
```

```python
Carpeta_json = 'C:\\Users\\gasty\\Desktop\\PF\\Google Maps\\metadata-sitios\\Faltante'
```

```python
def Cartepas_Df(carpeta):
    dataframes = []
    for archivo in os.listdir(carpeta):
        if archivo.endswith('.json'):
            ruta_completa = os.path.join(carpeta, archivo)
            try:
                df = pd.read_json(ruta_completa, lines=True)
                dataframes.append(df)
            except Exception as e:
                print(f"Error al cargar el archivo {archivo}: {e}")
    df_final = pd.concat(dataframes, ignore_index=True)
    return df_final
```

```python
metadata_8_11 = Cartepas_Df(Carpeta_json)
```
