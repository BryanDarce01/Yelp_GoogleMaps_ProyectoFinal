```python
import pandas as pd
import os
```

```python
capeta_data = 'C:\\Users\\gasty\\Desktop\\Proyecto Final Repo\\Yelp_GoogleMaps_ProyectoFinal\\Datasets\\metadata-sitios'
```

```python
def Cartepas_Df(carpeta):
    dataframes = []
    for archivo in os.listdir(carpeta):
        if archivo.endswith('.json'):
            ruta_completa = os.path.join(carpeta, archivo)
            try:
                df = pd.read_json(ruta_completa, lines=True)
                dataframes.append(df)
            except Exception as e:
                print(f"Error al cargar el archivo {archivo}: {e}")
    df_final = pd.concat(dataframes, ignore_index=True)
    return df_final
```

```python
metadata = Cartepas_Df(capeta_data)
```

```python
def extract_from_list(lst):
    if isinstance(lst, list) and len(lst) > 0:
        return ', '.join(lst)
    return None
```

```python
metadata.drop(columns=['price', 'description', 'MISC', 'relative_results', 'state', 'url'], inplace=True)
```

```python
metadata['category'] = metadata['category'].apply(extract_from_list)
```

```python
metadata['Monday'] = None
metadata['Tuesday'] = None
metadata['Wednesday'] = None
metadata['Thursday'] = None
metadata['Friday'] = None
metadata['Saturday'] = None
metadata['Sunday'] = None
```

```python
def extract_hours(row):
    if row['hours'] is not None:
        for day, hours in row['hours']:
            row[day] = hours
    return row
```

```python
metadata = metadata.apply(extract_hours, axis=1)
```

```python
metadata[['gmap_id1', 'gmap_id2']] = metadata['gmap_id'].str.split(':', expand=True)
```

```python
metadata.drop(columns=['gmap_id', 'hours'], inplace=True)
```

```python
metadata['name'] = metadata['name'].fillna('')
metadata['address'] = metadata['address'].fillna('')
metadata['category'] = metadata['category'].fillna('')
metadata['Monday'] = metadata['Monday'].fillna('')
metadata['Tuesday'] = metadata['Tuesday'].fillna('')
metadata['Wednesday'] = metadata['Wednesday'].fillna('')
metadata['Thursday'] = metadata['Thursday'].fillna('')
metadata['Friday'] = metadata['Friday'].fillna('')
metadata['Saturday'] = metadata['Saturday'].fillna('')
metadata['Sunday'] = metadata['Sunday'].fillna('')
```
