```python
import pandas as pd
import os
import json
```

# Texas

```python
ruta_archivo = r'Google Maps\reviews-estados\review-Texas\13.json'
```

```python
datos = []
with open(ruta_archivo, 'r') as archivo:
    for linea in archivo:
        objeto_json = json.loads(linea)
        datos.append(objeto_json)
```

```python
df1 = pd.DataFrame(datos)
```

```python
df1
```

```python
# Mostrar información básica del DataFrame
print("Información básica del DataFrame:")
print(df1.info())
```

```python
# Mostrar las primeras filas del DataFrame
print("\nPrimeras filas del DataFrame:")
print(df1.head())
```

```python
# Convertir la columna 'time' a tipo datetime y redondear al segundo más cercano
df1['time'] = pd.to_datetime(df1['time']).dt.round('1s')
```

```python
# Separar la columna 'time' en dos columnas: 'Date' y 'Time'
df1['Date'] = df1['time'].dt.date
df1['Time'] = df1['time'].dt.time
```

```python
# Convertir la columna 'rating' a tipo numérico (si es necesario)
df1['rating'] = pd.to_numeric(df1['rating'], errors='coerce')
```

```python
# eliminamos la columna original 'time' si ya no es necesaria
df1 = df1.drop('time', axis=1)
```

```python
# Verificar y eliminar filas con user_id nulos
df1 = df1.dropna(subset=['user_id'], axis=0)
```

```python
# Verificar y eliminar filas con user_id duplicados
df1 = df1.drop_duplicates(subset=['user_id'], keep='first')
```

```python
df1
```

Texas 2

```python
ruta_archivo2 = r'Google Maps\reviews-estados\review-Texas\16.json'
```

```python
datos2 = []
with open(ruta_archivo2, 'r') as archivo2:
    for linea in archivo2:
        objeto_json = json.loads(linea)
        datos2.append(objeto_json)

df2 = pd.DataFrame(datos2)
```

```python
# Convertir la columna 'time' a tipo datetime y redondear al segundo más cercano
df2['time'] = pd.to_datetime(df2['time']).dt.round('1s')
```

```python
# Convertir la columna 'rating' a tipo numérico (si es necesario)
df2['rating'] = pd.to_numeric(df2['rating'], errors='coerce')
```

```python
# Separar la columna 'time' en dos columnas: 'fecha' y 'hora'
df2['Date'] = df2['time'].dt.date
df2['Time'] = df2['time'].dt.time

# Verificar y eliminar filas con user_id nulos
df2 = df2.dropna(subset=['user_id'], axis=0)

# Verificar y eliminar filas con user_id duplicados
df2 = df2.drop_duplicates(subset=['user_id'], keep='first')
```

```python
# Concatenar los DataFrames df1 y df2
df_Texas = pd.concat([df1, df2], ignore_index=True)

# Llenar los valores nulos con NaN
df_Texas = df_Texas.fillna(pd.NA)
```

```python
df_Texas.to_csv('Google Maps\\reviews-estados\\df_texas.csv', index=False)
```

```python
# Lista con los nombres de Dataset a leer  en New York
archivos_ny = [
    '1.json', '2.json', '3.json', '4.json', '5.json', '6.json', '7.json', '8.json',
    '9.json', '10.json', '11.json', '12.json', '13.json', '14.json', '15.json', '16.json', '17.json', '18.json'
]

```

```python
# Lista para almacenar todos los datos de los archivos
datos_ny = []

# Leer y agregar los datos de cada archivo a la lista 'datos_ny'
for nombre_archivo in archivos_ny:
    ruta_archivo_ny = os.path.join('Google Maps', 'reviews-estados', 'review-New_York', nombre_archivo)
    with open(ruta_archivo_ny, 'r') as archivo_ny:
        for linea in archivo_ny:
            objeto_json = json.loads(linea)
            datos_ny.append(objeto_json)

# Crear el DataFrame 'df_ny' con los datos de todos los archivos
df_ny = pd.DataFrame(datos_ny)
```

```python
# Convertir la columna 'time' a tipo datetime y redondear al segundo más cercano
df_ny['time'] = pd.to_datetime(df_ny['time']).dt.round('1s')
```

```python
# Convertir la columna 'rating' a tipo numérico (si es necesario)
df_ny['rating'] = pd.to_numeric(df_ny['rating'], errors='coerce')
```

```python
# Separar la columna 'time' en dos columnas: 'fecha' y 'hora'
df_ny['Date'] = df_ny['time'].dt.date
df_ny['Time'] = df_ny['time'].dt.time
```
