1- Archivo business.pkl

Leemos el archivo

```python
df_business = pd.read_pickle('C:\\Users\\gasty\Desktop\\Proyecto Final Repo\\Yelp_GoogleMaps_ProyectoFinal\\Datasets\\Yelp\\business.pkl')
```

```python
print(df_business.shape)
df_business.head()
```

```python
print(df_business.isnull().sum())
```

```python
# Eliminar las columnas duplicadas con el mismo nombre
df_business = df_business.loc[:,~df_business.columns.duplicated()]
```

```python
df_business.at[0, 'state'] = 'CA'
```

```python
df_business.drop(columns=['attributes', 'is_open'], inplace=True)
```

```python
# Convertimos a númerico la columna
df_business['postal_code'] = pd.to_numeric(df_business['postal_code'], errors='coerce')
df_business['review_count'] = pd.to_numeric(df_business['review_count'], errors='coerce', downcast='integer')
df_business['latitude'] = pd.to_numeric(df_business['latitude'], errors='coerce', downcast='float')
df_business['longitude'] = pd.to_numeric(df_business['longitude'], errors='coerce', downcast='float')
df_business['stars'] = pd.to_numeric(df_business['stars'], errors='coerce', downcast='float')
```

```python
hours = pd.json_normalize(df_business['hours'])
```

```python
hours
```

```python
df_business = pd.concat([df_business.drop('hours', axis=1), hours], axis=1)
```

```python
df_business['postal_code'] = df_business['postal_code'].fillna(0)
df_business['categories'] = df_business['categories'].fillna('')
df_business['Monday'] = df_business['Monday'].fillna('')
df_business['Tuesday'] = df_business['Tuesday'].fillna('')
df_business['Wednesday'] = df_business['Wednesday'].fillna('')
df_business['Thursday'] = df_business['Thursday'].fillna('')
df_business['Friday'] = df_business['Friday'].fillna('')
df_business['Saturday'] = df_business['Saturday'].fillna('')
df_business['Sunday'] = df_business['Sunday'].fillna('')
```

```python
estados_deseados = ['CA', 'FL', 'NY']
```

```python
df_business = df_business[df_business['state'].isin(estados_deseados)]
```

```python
df_business = df_business.reset_index(drop=True)
```

```python
df_business.head()
```

2- Archivo 'review.json'

Leemos el archivo review.json de Yelp

```python
file_path = "C:\\Users\\gasty\\Desktop\\Proyecto Final Repo\\Yelp_GoogleMaps_ProyectoFinal\\Datasets\\Yelp\\review.json"

chunk_size = 10000
df_chunks = []


with open(file_path, 'r', encoding='utf-8') as f:
    for chunk in pd.read_json(f, lines=True, chunksize=chunk_size):
        df_chunks.append(chunk)


df_review = pd.concat(df_chunks, ignore_index=True)
```

```python
print(df_review.shape)
df_review.head()
```

```python
df_review.info()
```

```python
print(df_review.isnull().sum())
```

```python
df_review = df_review[df_review['date'].dt.year >= 2020]
```

```python
df_review.drop(columns=['useful', 'funny', 'cool'], inplace=True)
```

```python
df_review = df_review.reset_index(drop=True)
```
